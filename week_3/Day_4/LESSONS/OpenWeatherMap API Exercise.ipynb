```python
from IPython.display import clear_output as co
import json
import requests


class Weather:
    def __init__(self, city, condition, high, low, current, humidity):
        self.city = city
        self.condition = condition
        self.high = high
        self.low = low
        self.current = current
        self.humidity = humidity

    def get_city(self):
        return self.city

    def get_condition(self):
        return self.condition

    def get_high(self):
        return self.high

    def get_low(self):
        return self.low

    def get_current(self):
        return self.current
    
    def get_humidity(self):
        return self.humidity

    def __repr__(self):
        return f'<Weather: {self.high} {self.city}>'

    def __str__(self):
        return f'{self.high} {self.city}'

    # @classmethod
    def all(self):
        print("+:"*25)
        print(f'The weather in {(self.city).title()}')
        print(f'Current: {(self.current)}')
        print(f'With a high of: {(self.high)} and a low of {(self.low)}.')
        print(f'The humidity is: {(self.humidity)}.')
        print(":+"*25)


class WeatherDex:
    def __init__(self, city):
        self.web_link = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid=b5e21d8e54975d090069c0667cfd02d1'

    def run(self):
        url = requests.get(self.web_link)
        if url.status_code == 400:
            print('You must pass in a valid City Name to execute this program.')
        elif url.status_code == 200:
            data = url.json()
            condition = data['weather'][0]['description']
            high_temp = ((data['main']['temp_max'] - 273.15) * 9/5 + 32)
            high_temp = str(round(high_temp, 2))
            curr_temp = ((data['main']['temp'] - 273.15) * 9/5 + 32)
            curr_temp = str(round(curr_temp, 2))
            low_temp = ((data['main']['temp_min'] - 273.15) * 9/5 + 32)
            low_temp = str(round(low_temp, 2))
            humidity = data['main']['humidity']
            city = data['name']
            w = Weather(city,condition,high_temp,low_temp,curr_temp,humidity)
            w.all()

done = False
while not done:
    u_input = input("Choose a city to see its weather forecast, press q to quit: ")
    if u_input == 'q':
        done = True
    p = WeatherDex(u_input)
    p.run()
```
