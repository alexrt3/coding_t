##### FOR LOOP TO FIND ALL THE AVAILABLE DATES FOR EACH COUNTRY

```python
# This forloop takes the countries and adds them to a dictionary.  
# If it exists in the dict, then it will increase the value by +1

b_dict = dict()
country_list = ['United States', 'Ireland', 'Spain', 'Mexico', 'Canada', 'Singapore', 'Japan', 'United Kingdom', 'France']

# Holds the entire count data (including a key for each country)
for country in country_list:
    b_dict[country] = {}
    for person in data:
        # new_dict = b_dict[country] # This keeps track of the date counts for an individual country
        if person['country'] == country:
            # Loop: Print each date in a person's available dates
            available_dates = person['availableDates']
            for date in available_dates:
                if date in b_dict[country]:
                    b_dict[country][date] += 1
                else:
                    b_dict[country][date] = 1

print('United States')
print(b_dict['United States'])
print('Ireland')
print(b_dict['Ireland'])

# Next step is to get the highest value for each country. 
```
