```python
import requests
import json 

res = requests.get('http://ergast.com/api/f1/2008/driverStandings.json')
data = res.json()

DriverStandings = data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']

for i in range(22):
    print('\n' + DriverStandings[i]['Driver']['givenName'], DriverStandings[i]['Driver']['familyName'])
    print('Position:', DriverStandings[i]['position'])
    print('Date of Birth:', DriverStandings[i]['Driver']['dateOfBirth'])
    print('Number of Points:', DriverStandings[i]['points'])
    print('Nationality:', DriverStandings[i]['Driver']['nationality'])
    print('Constructor:', DriverStandings[i]['Constructors'][0]['name'] + '\n')
```
